## Query (Q), Key (K), and Value (V)

After tokenization, embedding lookup, positional encoding, and normalization, each token is represented by a dense vector that captures both its semantic meaning and positional information.

At this stage, however, each token is still independent. For example, in the sentence:

> **"cat sat on mat"**

the token **"cat"** knows only its own representation, **"sat"** knows only its own representation, and no interaction has occurred between them. The model has not yet learned how the tokens relate to one another.

For the Transformer to understand the context of a sentence, tokens must communicate with each other. They need a mechanism to determine:

- Which tokens are relevant?
- How relevant are they?
- What information should be exchanged?

A natural question arises:

**Why can't the Transformer compute attention directly on the embedding matrix?**

Although embeddings capture the semantic meaning of tokens, they are not designed to perform the different roles required by the attention mechanism. A single embedding cannot simultaneously represent:

- What information a token is searching for.
- What information a token contains.
- What information a token should pass to other tokens.

To solve this problem, the Transformer projects every input embedding into three different vector spaces using three learnable weight matrices.

$$
Q = XW_Q
$$

$$
K = XW_K
$$

$$
V = XW_V
$$

where:

- **Query (Q)** represents what information a token is searching for.
- **Key (K)** represents what information a token contains.
- **Value (V)** represents the information that will be passed to other tokens.

These three projections form the foundation of the self-attention mechanism. By comparing Query vectors with Key vectors, the Transformer determines how much attention each token should pay to every other token. The resulting attention weights are then applied to the Value vectors to produce context-aware representations of the input sequence.

After learning tokenization, embeddings, positional encoding, and normalization, the next step is to understand how the Transformer enables tokens to interact with one another.

The attention mechanism begins by projecting the input embeddings into three different vector spaces called **Query (Q)**, **Key (K)**, and **Value (V)**.

### 4. Matrix Multiplication

Using the example sentence:

> **"cat sat on mat"**

Take the input embedding matrix:

$$
X=
\begin{bmatrix}
0.3 & 0.8 & 0.1 & 0.5 \\
0.9 & 0.4 & 0.4 & 0.1 \\
0.5 & 0.2 & 0.8 & 0.3 \\
0.6 & 0.9 & 0.2 & 0.5
\end{bmatrix}
$$

Manually compute:

$$
Q = XW_Q
$$

Then compute:

$$
K = XW_K
$$

Finally compute:

$$
V = XW_V
$$

This step demonstrates how the same input embeddings are transformed into three different learned representations.

### 5. Attention Score Matrix

Once the Query and Key matrices are obtained, compute:

$$
QK^T
$$

This produces an attention score for every Query-Key pair.

Each element represents how strongly one token attends to another.

### 6. Scaled Dot Product

To stabilize the attention scores, divide by:

$$
\sqrt{d_k}
$$

Formula:

$$
\frac{QK^T}{\sqrt{d_k}}
$$

Understand why scaling is necessary before applying Softmax.

### 7. Softmax

Convert the scaled attention scores into probabilities:

$$
Softmax
\left(
\frac{QK^T}{\sqrt{d_k}}
\right)
$$

The resulting values indicate the attention weight assigned to each token.

### 8. Weighted Sum of Values

Multiply the attention weights by the Value matrix:

$$
Softmax
\left(
\frac{QK^T}{\sqrt{d_k}}
\right)V
$$

This produces a new contextual representation for every token.

### 9. Complete Scaled Dot Product Attention

Finally derive and understand the complete attention equation:

$$
Attention(Q,K,V)
=
Softmax
\left(
\frac{QK^T}{\sqrt{d_k}}
\right)V
$$

You should be able to explain the purpose of every matrix and every operation in this equation.

### 10. Multi-Head Attention

After mastering single-head attention, proceed to Multi-Head Attention.

Study:

$$
head_i
=
Attention(Q_i,K_i,V_i)
$$

Then combine all attention heads:

$$
MultiHead
=
Concat(head_1,\ldots,head_h)W_O
$$

Understand why multiple attention heads allow the model to capture different types of relationships simultaneously.

## Query (Q)

The **Query (Q)** is one of the three fundamental components of the self-attention mechanism in a Transformer. It represents the information that a token is searching for from all other tokens in the input sequence.

In simple terms, a Query can be thought of as a **question** asked by a token.

For example, consider the sentence:

> **"The cat sat on the mat."**

When processing the token **"sat"**, the Query generated for **"sat"** asks:

- Who performed the action?
- Which token is most relevant to me?
- Which tokens should I pay attention to?

The Query itself does not contain the answer. Instead, it is compared with the Key vectors of every token in the sequence to determine which tokens are most relevant.

### Why Do We Need Query?

If attention were computed directly on embeddings, every token would have only one representation.

However, attention requires every token to perform three different tasks:

- Search for relevant information.
- Describe the information it contains.
- Pass useful information to other tokens.

These tasks are fundamentally different.

Therefore, the Transformer learns three different representations for every token:

- Query (Q)
- Key (K)
- Value (V)

The Query is responsible only for searching.

### Mathematical Formulation

Suppose the input embedding matrix after positional encoding is:

$$
X
=
\begin{bmatrix}
0.3 & 0.8 & 0.1 & 0.5 \\
0.9 & 0.4 & 0.4 & 0.1 \\
0.5 & 0.2 & 0.8 & 0.3 \\
0.6 & 0.9 & 0.2 & 0.5
\end{bmatrix}
$$

The Query matrix is computed using a learnable weight matrix:

$$
Q=XW_Q
$$

where:

- \(X\) is the input embedding matrix.
- \(W_Q\) is the Query weight matrix.
- \(Q\) is the Query matrix.

### Dimensions

Suppose:

- Sequence Length = 4
- Embedding Dimension = 4

Then,

$$
X
\in
\mathbb{R}^{4\times4}
$$

If

$$
W_Q
\in
\mathbb{R}^{4\times2}
$$

then

$$
Q
=
XW_Q
\in
\mathbb{R}^{4\times2}
$$

Each row of \(Q\) corresponds to the Query vector of one token.

### How is \(W_Q\) Initialized?

The Query weight matrix is initialized randomly before training.

Typical initialization methods include:

- Xavier Initialization
- Kaiming Initialization
- Normal Distribution

Example:

$$
W_Q=
\begin{bmatrix}
0.12 & -0.08\\
0.04 & 0.17\\
-0.09 & 0.11\\
0.21 & -0.03
\end{bmatrix}
$$

Initially, these values have no semantic meaning.

During training, backpropagation updates \(W_Q\) so that the Query vectors learn how to search for relevant information.

### Example Computation

Suppose

$$
W_Q=
\begin{bmatrix}
1 & 0\\
0 & 1\\
1 & 0\\
0 & 1
\end{bmatrix}
$$

Then,

$$
Q=XW_Q
$$

becomes

$$
Q=
\begin{bmatrix}
0.4 & 1.3\\
1.3 & 0.5\\
1.3 & 0.5\\
0.8 & 1.4
\end{bmatrix}
$$

Each row now represents the Query vector for one token.

### Interpretation

For the sentence:

> **"cat sat on mat"**

The Query vector for:

**cat**

asks:

> Which tokens are important for understanding "cat"?

The Query vector for:

**sat**

asks:

> Which token tells me who performed the action?

The Query vector for:

**mat**

asks:

> Which tokens describe my relationship with the sentence?

Each Query is different because each token has a different role in the sentence.

### What Happens Next?

The Query vectors are compared with every Key vector using the dot product.

Mathematically,

$$
Score
=
QK^T
$$

Each element of this matrix represents how strongly a Query matches a Key.

These scores are later converted into probabilities using the Softmax function.

### Important Points to Remember

- Every token has exactly one Query vector.
- Query vectors are learned during training.
- Query vectors determine what information a token is searching for.
- Query vectors are computed using a linear transformation of the input embeddings.
- Query vectors are compared with Key vectors to compute attention scores.
- Query vectors never directly interact with Value vectors.
- The Query matrix has the same number of rows as the number of input tokens.
- Query vectors are recomputed at every Transformer layer.
- Different Transformer layers learn different Query representations.
- Without Query vectors, the attention mechanism cannot determine which information each token should retrieve.

### Interview Question

**What is the purpose of the Query vector in a Transformer?**

**Answer:**

The Query vector represents what information a token is searching for. It is obtained by projecting the input embeddings using a learnable weight matrix \(W_Q\). During self-attention, each Query vector is compared with every Key vector to compute attention scores, which determine how much attention the token should pay to other tokens in the sequence.
