
# Parameter-Efficient Fine-Tuning (PEFT)

## Pre-Context

Large Language Models (LLMs) are initially trained on massive text corpora using self-supervised learning, enabling them to acquire broad linguistic knowledge and reasoning capabilities. Although these pretrained models perform well on general tasks, they often require further adaptation to specialized domains such as medicine, finance, law, cybersecurity, scientific research, and software engineering.

Traditionally, this adaptation was performed using **Full Fine-Tuning (FFT)**, where every parameter of the pretrained model is updated. While effective, Full Fine-Tuning becomes increasingly impractical as model size grows. Modern LLMs contain billions or even trillions of parameters, making full optimization extremely expensive in terms of GPU memory, computational resources, storage, and training time.

To address these limitations, researchers introduced **Parameter-Efficient Fine-Tuning (PEFT)**. Instead of updating the entire network, PEFT freezes the pretrained model and trains only a small number of additional parameters. This significantly reduces memory consumption while maintaining performance close to Full Fine-Tuning.

Today, PEFT has become the standard approach for adapting modern LLMs because it enables efficient fine-tuning on commodity hardware without sacrificing model quality.

---

# What is Parameter-Efficient Fine-Tuning (PEFT)?

**Parameter-Efficient Fine-Tuning (PEFT)** is a family of techniques designed to adapt pretrained neural networks by updating only a small subset of trainable parameters while keeping the majority of the pretrained model frozen.

Unlike Full Fine-Tuning, which modifies every parameter in the model, PEFT introduces lightweight trainable components or selectively updates existing parameters. The objective is to preserve the general knowledge learned during pretraining while learning only the task-specific information required for a downstream application.

Mathematically, let the pretrained model parameters be represented as

$$
\theta
=
\{\theta_1,\theta_2,\ldots,\theta_N\}
$$

where

- \(N\) denotes the total number of model parameters.

In Full Fine-Tuning, every parameter is updated:

$$
\theta
\leftarrow
\theta
-
\eta
\nabla_{\theta}
\mathcal{L}
$$

where

- \(\eta\) is the learning rate.
- \(\mathcal{L}\) is the training loss.

In PEFT, the parameter set is divided into two groups:

$$
\theta
=
\theta_{\text{frozen}}
\cup
\phi_{\text{trainable}}
$$

where

- \(\theta_{\text{frozen}}\) represents the frozen pretrained parameters.
- \(\phi_{\text{trainable}}\) represents the newly introduced or selectively trainable parameters.

The defining characteristic of PEFT is

$$
|\phi_{\text{trainable}}|
\ll
|\theta_{\text{frozen}}|
$$

meaning that only a tiny fraction of the model is optimized.

---

# Why Do We Need PEFT?

The motivation for PEFT arises from several practical limitations of Full Fine-Tuning.

### 1. Extremely Large Models

Modern language models contain billions of parameters. Updating every parameter requires substantial computational resources.

### 2. GPU Memory Constraints

During training, memory is required not only for model weights but also for gradients, optimizer states, and intermediate activations. Full Fine-Tuning often exceeds the memory capacity of consumer GPUs.

### 3. Storage Cost

A separate fully fine-tuned checkpoint must be stored for every downstream task. Maintaining multiple copies of billion-parameter models quickly becomes prohibitively expensive.

### 4. Training Time

Updating billions of parameters significantly increases computational cost and training duration.

### 5. Multi-Task Adaptation

Organizations often need to adapt a single pretrained model to many different applications. PEFT allows each task to store only lightweight adapters instead of duplicating the entire model.

---

# Advantages of PEFT

Parameter-Efficient Fine-Tuning provides several practical advantages over Full Fine-Tuning.

- Reduces the number of trainable parameters.
- Lowers GPU memory consumption.
- Reduces optimizer memory.
- Decreases checkpoint storage requirements.
- Accelerates training.
- Enables fine-tuning on consumer hardware.
- Supports multiple task-specific adapters using a shared pretrained model.
- Often achieves performance comparable to Full Fine-Tuning.

---

# General PEFT Framework

Most PEFT methods follow the same high-level workflow.

$$
\boxed{
\text{Pretrained Model}
\rightarrow
\text{Freeze Base Parameters}
\rightarrow
\text{Insert Trainable Components}
\rightarrow
\text{Fine-Tune}
\rightarrow
\text{Save Lightweight Parameters}
}
$$

Instead of modifying the entire model, only the inserted components or selected parameters are updated.

---

# Types of Parameter-Efficient Fine-Tuning (PEFT)

PEFT is not a single algorithm. It is a collection of techniques that achieve parameter-efficient adaptation using different strategies.

The major categories include:

---

## 1. Adapter-Based Methods

Adapter methods insert small neural network modules between existing Transformer layers.

Rather than modifying the original weights, the adapter learns task-specific transformations while the pretrained model remains frozen.

Examples:

- Houlsby Adapters
- Pfeiffer Adapters
- Compacter

---

## 2. Low-Rank Adaptation Methods

These methods approximate weight updates using low-rank matrix decomposition.

Instead of learning the full update matrix, they learn two smaller matrices whose product approximates the desired weight update.

Examples:

- LoRA
- AdaLoRA
- DoRA
- LoRA+
- VeRA
- LoHa
- LoKr

---

## 3. Prompt-Based Methods

Rather than modifying model weights, these methods learn trainable prompt vectors that are prepended to the input or intermediate representations.

Examples:

- Prompt Tuning
- Prefix Tuning
- P-Tuning
- P-Tuning v2

---

## 4. Activation Scaling Methods

Instead of introducing new layers, these approaches learn scaling coefficients that modulate existing activations within the Transformer.

Examples:

- IA³

---

## 5. Quantization-Based PEFT

These methods combine parameter-efficient fine-tuning with model quantization to further reduce GPU memory requirements.

Examples:

- QLoRA

---

## 6. Hybrid PEFT Methods

Hybrid approaches combine multiple PEFT techniques to achieve improved performance or greater memory efficiency.

Examples:

- LoRA + Prefix Tuning
- LoRA + Adapters
- Quantized Adapters
- Sparse LoRA

---

# Comparison of Major PEFT Categories

| Category | Main Idea | Example Methods |
|-----------|-----------|-----------------|
| Adapter-Based | Insert trainable neural modules | Houlsby Adapter, Pfeiffer Adapter |
| Low-Rank Adaptation | Learn low-rank weight updates | LoRA, AdaLoRA, DoRA |

| Quantization-Based | Combine PEFT with quantization | QLoRA |
| Hybrid Methods | Combine multiple PEFT approaches | LoRA + Prefix, Sparse LoRA |

# Low-Rank Adaptation (LoRA)

## Pre-Context

Large Language Models (LLMs) such as GPT, Llama, Gemma, Mistral, and Qwen contain billions of trainable parameters. Traditionally, adapting these models to a new task required **Full Fine-Tuning (FFT)**, where every parameter of the model is updated using task-specific data.

Although Full Fine-Tuning achieves excellent performance, it is extremely expensive in terms of GPU memory, computational resources, storage, and training time. For example, fine-tuning a 70B parameter model requires hundreds of gigabytes of GPU memory, making it inaccessible for most researchers and organizations.

To overcome these limitations, researchers introduced **Parameter-Efficient Fine-Tuning (PEFT)** methods. Instead of updating every parameter, PEFT methods modify only a very small subset of trainable parameters while keeping the original model frozen.

One of the most influential PEFT techniques is **Low-Rank Adaptation (LoRA)**, introduced by Microsoft Research in 2021. LoRA enables efficient fine-tuning by learning small low-rank matrices instead of updating the entire weight matrix, reducing memory requirements by orders of magnitude while maintaining performance close to Full Fine-Tuning.

---

# What is LoRA?

**Low-Rank Adaptation (LoRA)** is a Parameter-Efficient Fine-Tuning (PEFT) technique that adapts a pretrained Large Language Model by freezing its original weights and learning only a pair of small trainable matrices.

Instead of modifying the original weight matrix of a neural network layer, LoRA assumes that the required weight update can be approximated using a **low-rank decomposition**. This dramatically reduces the number of trainable parameters while preserving most of the model's performance.

Rather than updating

$$
W
$$

LoRA learns only

$$
\Delta W
$$

where

$$
\Delta W = BA
$$

and

$$
r \ll d
$$

where \(r\) is the low-rank dimension.

The original pretrained model remains unchanged throughout training.

---

# Why Do We Need LoRA?

Full Fine-Tuning has several limitations:

- Every model parameter must be updated.
- GPU memory requirements are extremely high.
- Optimizer states consume significant memory.
- Fine-tuning large models becomes prohibitively expensive.
- A separate copy of the entire model must be stored for every downstream task.

LoRA addresses these challenges by introducing a lightweight adaptation mechanism.

Instead of updating billions of parameters, LoRA updates only a few million parameters, making fine-tuning significantly more memory-efficient and computationally cheaper.

---

# Why Does Low-Rank Approximation Work?

The key insight behind LoRA is that task-specific weight updates often occupy only a **small subspace** of the full parameter space.

Although the original weight matrix may contain millions of parameters, the changes required for adapting the model to a new task are often highly structured and can be represented using a much lower-dimensional matrix.

Instead of learning an arbitrary matrix

$$
\Delta W
\in
\mathbb{R}^{d \times k}
$$

LoRA approximates it as

$$
\Delta W
=
BA
$$

where

$$
B
\in
\mathbb{R}^{d\times r}
$$

and

$$
A
\in
\mathbb{R}^{r\times k}
$$

with

$$
r
\ll
d,k
$$

This drastically reduces the number of trainable parameters.

---

# Full Fine-Tuning vs LoRA

Suppose a Transformer projection layer contains

$$
W
\in
\mathbb{R}^{4096\times4096}
$$

Total parameters:

$$
4096\times4096
=
16,777,216
$$

Full Fine-Tuning updates all

$$
16.8
\text{ million parameters}
$$

Now assume

$$
r=16
$$

LoRA learns

$$
A
\in
\mathbb{R}^{16\times4096}
$$

and

$$
B
\in
\mathbb{R}^{4096\times16}
$$

Total trainable parameters:

$$
16\times4096
+
4096\times16
=
131,072
$$

Instead of updating

$$
16.8
\text{ million}
$$

parameters, LoRA updates only

$$
131
\text{ thousand}
$$

which is approximately

$$
99.2\%
$$

fewer parameters.

---

# Mathematical Formulation

Assume the original linear layer is

$$
y
=
Wx
$$

where

- \(W\) is the pretrained weight matrix.
- \(x\) is the input.
- \(y\) is the output.

During Full Fine-Tuning,

$$
W
\rightarrow
W+\Delta W
$$

where

$$
\Delta W
$$

contains millions of trainable parameters.

LoRA replaces this update with

$$
\Delta W
=
BA
$$

Therefore,

$$
W'
=
W
+
BA
$$

The forward pass becomes

$$
y
=
(W+BA)x
$$

Since

$$
W
$$

is frozen,

only

$$
A
$$

and

$$
B
$$

are updated during backpropagation.

---

# Matrix Dimensions

Suppose

$$
W
\in
\mathbb{R}^{d\times k}
$$

Then

$$
A
\in
\mathbb{R}^{r\times k}
$$

$$
B
\in
\mathbb{R}^{d\times r}
$$

where

- \(d\) = output dimension
- \(k\) = input dimension
- \(r\) = LoRA rank

Since

$$
r
\ll
d,k
$$

the number of trainable parameters becomes

$$
r(d+k)
$$

instead of

$$
dk
$$

---

# Forward Pass

The computation inside a LoRA layer is

$$
Input
\rightarrow
Wx
\rightarrow
BAx
\rightarrow
Addition
\rightarrow
Output
$$

Mathematically,

$$
y
=
Wx
+
BAx
$$

The pretrained weights provide the original knowledge,

while

$$
BA
$$

learns the task-specific adaptation.

---

# Backpropagation

During training,

the gradients are computed only for

$$
A
$$

and

$$
B
$$

The original matrix

$$
W
$$

remains frozen.

Therefore,

$$
\frac{\partial L}{\partial W}
=
0
$$

while

$$
\frac{\partial L}{\partial A}
\neq
0
$$

and

$$
\frac{\partial L}{\partial B}
\neq
0
$$

This significantly reduces optimizer memory usage because gradients and optimizer states are stored only for the LoRA matrices.

---

# LoRA Hyperparameters

## Rank (r)

The rank determines the size of the low-rank matrices.

Small values:

- 4
- 8
- 16

consume less memory.

Larger values increase adaptation capacity but require more trainable parameters.

---

## Alpha

LoRA scales the update using

$$
\frac{\alpha}{r}
$$

The adapted weight becomes

$$
W'
=
W
+
\frac{\alpha}{r}
BA
$$

Typical values:

- 8
- 16
- 32

---

## Dropout

LoRA optionally applies dropout to improve generalization during fine-tuning.

Typical values:

- 0
- 0.05
- 0.1

---

# Advantages

LoRA provides several advantages over Full Fine-Tuning:

- Updates only a small number of parameters.
- Significantly reduces GPU memory consumption.
- Reduces optimizer memory.
- Faster training.
- Smaller checkpoint sizes.
- Easy to share adapters.
- Multiple LoRA adapters can be trained for different tasks using the same base model.
- Performance is often comparable to Full Fine-Tuning.

---

# Limitations

Although LoRA is highly efficient, it has some limitations.

- Performance may slightly degrade for highly specialized tasks.
- Choosing an appropriate rank requires experimentation.
- Very low ranks may underfit complex tasks.
- The frozen base model cannot learn new general knowledge.
