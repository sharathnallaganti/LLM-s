# Is DPO the Same as Pretraining?

At first glance, **Pretraining** and **Direct Preference Optimization (DPO)** appear very similar because both use the same optimization pipeline:

$$
\text{Forward Pass}
\rightarrow
\text{Loss}
\rightarrow
\text{Backpropagation}
\rightarrow
\text{Weight Update}
$$

However, they differ in **what they learn**, **the training data they use**, and **the loss function they optimize**.

## Comparison

| Pretraining | Direct Preference Optimization (DPO) |
|-------------|--------------------------------------|
| Learns language patterns | Learns human preferences |
| Uses raw text datasets | Uses prompt, chosen response, and rejected response |
| Predicts the next token | Learns to prefer one response over another |
| Uses Cross Entropy Loss | Uses DPO Loss |
| Objective is language modeling | Objective is alignment |

## Pretraining Pipeline

During pretraining, the model learns to predict the next token from raw text.

Suppose the training sentence is

```text
The capital of France is Paris.
```

The model receives

```text
Input:
The capital of France is

Target:
Paris
```

The training pipeline is

$$
\text{Input}
\rightarrow
\text{Forward Pass}
\rightarrow
\text{Logits}
\rightarrow
\text{Cross Entropy Loss}
\rightarrow
\text{Backpropagation}
\rightarrow
\text{Update Weights}
$$

The model simply learns:

> "Given the previous tokens, predict the next token."

---

## DPO Pipeline

During DPO, the model no longer learns from raw text.

Instead, it learns from **human preference pairs**.

Suppose the prompt is

```text
What is the capital of France?
```

The model generates

**Response A**

```text
Paris
```

**Response B**

```text
London
```

Human annotators select

```text
✓ Paris
```

and reject

```text
✗ London
```

The training pipeline becomes

$$
\text{Prompt}
\rightarrow
\text{Generate Responses}
\rightarrow
\text{Chosen Response}
+
\text{Rejected Response}
\rightarrow
\text{DPO Loss}
\rightarrow
\text{Backpropagation}
\rightarrow
\text{Update Weights}
$$

Notice that the model is **not predicting the next token from a dataset**. Instead, it is learning which complete response humans prefer.

---

## What Actually Changes?

### During Pretraining

The model learns

$$
P(\text{Next Token}\mid\text{Previous Tokens})
$$

Example

```text
The capital of France is

↓

Paris
```

Loss Function

$$
\text{Cross Entropy Loss}
$$

---

### During DPO

The model learns

$$
P(\text{Chosen Response})
>
P(\text{Rejected Response})
$$

Example

```text
Prompt

↓

Paris

↓

London

↓

Prefer Paris
```

Loss Function

$$
\text{DPO Loss}
$$

---

## Similarity Between Pretraining and DPO

Both training methods use gradient-based optimization.

The optimization loop is identical:

$$
\text{Forward Pass}
\rightarrow
\text{Loss}
\rightarrow
\text{Backpropagation}
\rightarrow
\text{Gradient Descent}
\rightarrow
\text{Weight Update}
$$

The only difference is the objective being optimized.

- **Pretraining:** Optimize Cross Entropy Loss for next-token prediction.
- **DPO:** Optimize DPO Loss for human preference alignment.

---

## Simple Analogy

Imagine teaching a student.

### Stage 1: Pretraining

You give the student millions of books.

The student learns:

- Vocabulary
- Grammar
- Facts
- Language patterns

---

### Stage 2: DPO

Now you ask the student a question.

Question:

```text
Write a polite email.
```

The student writes two answers.

You simply say:

> "Answer A is better than Answer B."

After repeating this process thousands of times, the student learns **which responses humans prefer**, rather than learning new facts.

---

## Key Difference

**Pretraining teaches the model what language looks like.**

**Direct Preference Optimization teaches the model which responses humans prefer.**

Although both methods use the same optimization pipeline, they optimize different objectives and therefore serve different purposes in the training of a Large Language Model.

# Pretraining vs Direct Preference Optimization (DPO) Pipeline

Although both **Pretraining** and **Direct Preference Optimization (DPO)** update the model weights using gradient descent, they differ in the type of data they use, their learning objective, and the loss function they optimize.

## Pretraining Pipeline

During pretraining, the model learns to predict the next token from large-scale text corpora.

$$
\boxed{
\text{Raw Text}
\rightarrow
\text{Tokenization}
\rightarrow
\text{Token IDs}
\rightarrow
\text{Embedding Layer}
\rightarrow
\text{Transformer}
\rightarrow
\text{Logits}
\rightarrow
\text{Cross Entropy Loss}
\rightarrow
\text{Backpropagation}
\rightarrow
\text{Weight Update}
}
$$

### Example

Training Text:

```text
The capital of France is Paris.
```

Input:

```text
The capital of France is
```

Target:

```text
Paris
```

The objective is to maximize

$$
P(\text{Paris}\mid\text{The capital of France is})
$$

---

## Direct Preference Optimization (DPO) Pipeline

During DPO, the model learns from **human preference pairs** rather than raw text.

$$
\boxed{
\text{Prompt}
\rightarrow
\text{Generate Multiple Responses}
\rightarrow
\text{Chosen Response}
+
\text{Rejected Response}
\rightarrow
\text{Current Model}
+
\text{Reference Model}
\rightarrow
\text{DPO Loss}
\rightarrow
\text{Backpropagation}
\rightarrow
\text{Weight Update}
}
$$

### Example

Prompt:

```text
What is the capital of France?
```

Generated Responses:

```text
A: Paris
```

```text
B: London
```

Human Preference:

```text
✓ Paris
```

```text
✗ London
```

The objective is to make

$$
P(\text{Paris})
>
P(\text{London})
$$

---

# Pipeline Comparison

| Pretraining | Direct Preference Optimization (DPO) |
|-------------|--------------------------------------|
| Raw text dataset | Human preference dataset |
| Predict next token | Compare two complete responses |
| Uses one target answer | Uses chosen and rejected responses |
| Cross Entropy Loss | DPO Loss |
| Learns language patterns | Learns human preferences |
| No human feedback | Requires human preference annotations |
| Produces a base model | Produces an aligned model |

---

# Overall LLM Training Pipeline

The complete training process of a modern Large Language Model is

$$
\boxed{
\text{Raw Text}
\rightarrow
\text{Pretraining}
\rightarrow
\text{Base Model}
\rightarrow
\text{Supervised Fine-Tuning (SFT)}
\rightarrow
\text{Direct Preference Optimization (DPO)}
\rightarrow
\text{Aligned Large Language Model}
}
$$

## Key Difference

**Pretraining teaches the model _what language looks like_.**

**DPO teaches the model _which response humans prefer_.**

Although both use **forward pass → loss → backpropagation → weight update**, they optimize different objectives using different datasets.
