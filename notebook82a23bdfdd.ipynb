# Is DPO the Same as Pretraining?

At first glance, **Pretraining** and **Direct Preference Optimization (DPO)** appear very similar because both use the same optimization pipeline:

$$
\text{Forward Pass}
\rightarrow
\text{Loss}
\rightarrow
\text{Backpropagation}
\rightarrow
\text{Weight Update}
$$

However, they differ in **what they learn**, **the training data they use**, and **the loss function they optimize**.

## Comparison

| Pretraining | Direct Preference Optimization (DPO) |
|-------------|--------------------------------------|
| Learns language patterns | Learns human preferences |
| Uses raw text datasets | Uses prompt, chosen response, and rejected response |
| Predicts the next token | Learns to prefer one response over another |
| Uses Cross Entropy Loss | Uses DPO Loss |
| Objective is language modeling | Objective is alignment |

## Pretraining Pipeline

During pretraining, the model learns to predict the next token from raw text.

Suppose the training sentence is

```text
The capital of France is Paris.
```

The model receives

```text
Input:
The capital of France is

Target:
Paris
```

The training pipeline is

$$
\text{Input}
\rightarrow
\text{Forward Pass}
\rightarrow
\text{Logits}
\rightarrow
\text{Cross Entropy Loss}
\rightarrow
\text{Backpropagation}
\rightarrow
\text{Update Weights}
$$

The model simply learns:

> "Given the previous tokens, predict the next token."

---

## DPO Pipeline

During DPO, the model no longer learns from raw text.

Instead, it learns from **human preference pairs**.

Suppose the prompt is

```text
What is the capital of France?
```

The model generates

**Response A**

```text
Paris
```

**Response B**

```text
London
```

Human annotators select

```text
✓ Paris
```

and reject

```text
✗ London
```

The training pipeline becomes

$$
\text{Prompt}
\rightarrow
\text{Generate Responses}
\rightarrow
\text{Chosen Response}
+
\text{Rejected Response}
\rightarrow
\text{DPO Loss}
\rightarrow
\text{Backpropagation}
\rightarrow
\text{Update Weights}
$$

Notice that the model is **not predicting the next token from a dataset**. Instead, it is learning which complete response humans prefer.

---

## What Actually Changes?

### During Pretraining

The model learns

$$
P(\text{Next Token}\mid\text{Previous Tokens})
$$

Example

```text
The capital of France is

↓

Paris
```

Loss Function

$$
\text{Cross Entropy Loss}
$$

---

### During DPO

The model learns

$$
P(\text{Chosen Response})
>
P(\text{Rejected Response})
$$

Example

```text
Prompt

↓

Paris

↓

London

↓

Prefer Paris
```

Loss Function

$$
\text{DPO Loss}
$$

---

## Similarity Between Pretraining and DPO

Both training methods use gradient-based optimization.

The optimization loop is identical:

$$
\text{Forward Pass}
\rightarrow
\text{Loss}
\rightarrow
\text{Backpropagation}
\rightarrow
\text{Gradient Descent}
\rightarrow
\text{Weight Update}
$$

The only difference is the objective being optimized.

- **Pretraining:** Optimize Cross Entropy Loss for next-token prediction.
- **DPO:** Optimize DPO Loss for human preference alignment.

---

## Simple Analogy

Imagine teaching a student.

### Stage 1: Pretraining

You give the student millions of books.

The student learns:

- Vocabulary
- Grammar
- Facts
- Language patterns

---

### Stage 2: DPO

Now you ask the student a question.

Question:

```text
Write a polite email.
```

The student writes two answers.

You simply say:

> "Answer A is better than Answer B."

After repeating this process thousands of times, the student learns **which responses humans prefer**, rather than learning new facts.

---

## Key Difference

**Pretraining teaches the model what language looks like.**

**Direct Preference Optimization teaches the model which responses humans prefer.**

Although both methods use the same optimization pipeline, they optimize different objectives and therefore serve different purposes in the training of a Large Language Model.
