# AI and LLM Acronyms Reference

This notebook summarizes important concepts used in large language models,
fine-tuning, alignment, retrieval, optimization, and transformer architectures.

---

## 1. Large Language Models

### LLM — Large Language Model

A large language model estimates the probability of a sequence of tokens:

$$
P(x_1, x_2, \dots, x_n)
=
\prod_{t=1}^{n}
P(x_t \mid x_1, x_2, \dots, x_{t-1})
$$

The model predicts the next token using all previous tokens as context.

---

## 2. LoRA — Low-Rank Adaptation

LoRA fine-tunes a model by adding a low-rank update to a pretrained weight matrix.

Let the original weight matrix be:

$$
W_0 \in \mathbb{R}^{d \times k}
$$

Instead of updating all parameters in $W_0$, LoRA learns:

$$
\Delta W = BA
$$

where:

$$
B \in \mathbb{R}^{d \times r}
$$

and

$$
A \in \mathbb{R}^{r \times k}
$$

The adapted weight becomes:

$$
W = W_0 + \frac{\alpha}{r}BA
$$

where:

- $r$ is the LoRA rank
- $\alpha$ is the scaling factor
- $W_0$ remains frozen
- $A$ and $B$ are trainable

### Why LoRA is useful

- Reduces trainable parameters
- Requires less GPU memory
- Produces small adapter files
- Allows multiple task-specific adapters

---

## 3. QLoRA — Quantized Low-Rank Adaptation

QLoRA combines LoRA with quantization.

The base model weights are stored using low precision:

$$
W_q = Q(W_0)
$$

The effective adapted weight is:

$$
W = \operatorname{dequantize}(W_q)
+
\frac{\alpha}{r}BA
$$

QLoRA commonly uses 4-bit NormalFloat quantization.

### NF4 — NormalFloat 4-bit

NF4 is designed for neural-network weights that approximately follow a normal
distribution.

QLoRA reduces memory usage while preserving strong fine-tuning performance.

---

## 4. DoRA — Weight-Decomposed Low-Rank Adaptation

DoRA separates a weight matrix into magnitude and direction.

A weight vector can be written as:

$$
W = m \frac{V}{\|V\|}
$$

where:

- $m$ controls magnitude
- $V$ controls direction

DoRA applies low-rank adaptation primarily to the directional component.

---

## 5. SVD — Singular Value Decomposition

For a matrix:

$$
W \in \mathbb{R}^{m \times n}
$$

SVD factorizes it as:

$$
W = U\Sigma V^\top
$$

where:

- $U$ contains left singular vectors
- $\Sigma$ contains singular values
- $V$ contains right singular vectors

A rank-$r$ approximation is:

$$
W_r = U_r \Sigma_r V_r^\top
$$

SVD helps explain why low-rank methods such as LoRA can work effectively.

---

## 6. Transformer Attention

Scaled dot-product attention is:

$$
\operatorname{Attention}(Q,K,V)
=
\operatorname{softmax}
\left(
\frac{QK^\top}{\sqrt{d_k}}
\right)V
$$

where:

- $Q$ represents queries
- $K$ represents keys
- $V$ represents values
- $d_k$ is the key dimension

The scaling factor:

$$
\frac{1}{\sqrt{d_k}}
$$

helps prevent very large dot products.

---

## 7. MHA — Multi-Head Attention

Each attention head is:

$$
\operatorname{head}_i
=
\operatorname{Attention}
\left(
QW_i^Q,
KW_i^K,
VW_i^V
\right)
$$

The heads are combined using:

$$
\operatorname{MHA}(Q,K,V)
=
\operatorname{Concat}
\left(
\operatorname{head}_1,
\dots,
\operatorname{head}_h
\right)
W^O
$$

---

## 8. GQA — Grouped Query Attention

Grouped Query Attention shares key and value projections across groups of query
heads.

It provides a compromise between:

- Multi-Head Attention
- Multi-Query Attention

Its main benefit is reduced inference memory and faster token generation.

---

## 9. Cross-Entropy Loss

For a target distribution $y$ and prediction $\hat{y}$:

$$
\mathcal{L}_{CE}
=
-\sum_{i=1}^{C}
y_i \log(\hat{y}_i)
$$

For next-token prediction:

$$
\mathcal{L}
=
-\sum_{t=1}^{T}
\log
P_\theta
\left(
x_t \mid x_{<t}
\right)
$$

---

## 10. DPO — Direct Preference Optimization

DPO trains a model using preferred and rejected responses.

Let:

- $y_w$ be the preferred response
- $y_l$ be the rejected response
- $\pi_\theta$ be the trainable policy
- $\pi_{\mathrm{ref}}$ be the reference policy

The DPO objective is:

$$
\mathcal{L}_{DPO}
=
-
\log
\sigma
\left[
\beta
\left(
\log
\frac{\pi_\theta(y_w \mid x)}
{\pi_{\mathrm{ref}}(y_w \mid x)}
-
\log
\frac{\pi_\theta(y_l \mid x)}
{\pi_{\mathrm{ref}}(y_l \mid x)}
\right)
\right]
$$

where $\sigma$ is the sigmoid function and $\beta$ controls preference strength.

---

## 11. ORPO — Odds Ratio Preference Optimization

ORPO combines supervised fine-tuning with preference optimization.

Its objective can be represented as:

$$
\mathcal{L}_{ORPO}
=
\mathcal{L}_{SFT}
+
\lambda
\mathcal{L}_{OR}
$$

The odds of a response are:

$$
\operatorname{odds}(y \mid x)
=
\frac{P(y \mid x)}
{1-P(y \mid x)}
$$

The preference term encourages the preferred response to have higher odds than
the rejected response.

---

## 12. PPO — Proximal Policy Optimization

PPO uses a clipped objective:

$$
L^{CLIP}(\theta)
=
\mathbb{E}_t
\left[
\min
\left(
r_t(\theta)A_t,
\operatorname{clip}
\left(
r_t(\theta),
1-\epsilon,
1+\epsilon
\right)
A_t
\right)
\right]
$$

where:

$$
r_t(\theta)
=
\frac{
\pi_\theta(a_t \mid s_t)
}{
\pi_{\theta_{\mathrm{old}}}(a_t \mid s_t)
}
$$

PPO is widely associated with reinforcement learning from human feedback.

---

## 13. RLHF — Reinforcement Learning from Human Feedback

A simplified RLHF pipeline contains:

1. Supervised fine-tuning
2. Human preference collection
3. Reward-model training
4. Policy optimization

A reward model may learn:

$$
R_\phi(x,y)
$$

The policy is optimized to maximize reward while remaining close to a reference
policy:

$$
\max_\theta
\mathbb{E}
\left[
R_\phi(x,y)
-
\beta
D_{KL}
\left(
\pi_\theta
\|
\pi_{\mathrm{ref}}
\right)
\right]
$$

---

## 14. KL Divergence

For distributions $P$ and $Q$:

$$
D_{KL}(P\|Q)
=
\sum_x
P(x)
\log
\frac{P(x)}{Q(x)}
$$

KL divergence measures how one probability distribution differs from another.

---

## 15. RAG — Retrieval-Augmented Generation

A RAG system retrieves documents before generating an answer.

The retrieval stage finds:

$$
D_k
=
\operatorname{TopK}
\left(
\operatorname{sim}(q,d_i)
\right)
$$

The generator produces:

$$
P(y \mid q,D_k)
$$
